1. import & data load

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = sns.load_dataset("iris")
data.head()
```

```python
# X = feature values,  all the columns except last
X = data.iloc[:,:-1]
# y = target values, last column
y = data.iloc[:,-1]
```

2. PLOT

```python
plt.xlabel('features')
plt.ylabel('species')

pltX = data.loc[:, 'sepal_length']
pltY = data.loc[:, 'species']
plt.scatter(pltX, pltY, color = 'blue', label = 'sepal_length')

pltX = data.loc[:, 'sepal_width']
pltY = data.loc[:, 'species']
plt.scatter(pltX, pltY, color = 'green', label = 'sepal_width')

pltX = data.loc[:, 'petal_length']
pltY = data.loc[:, 'species']
plt.scatter(pltX, pltY, color = 'orange', label = 'petal_length')

pltX = data.loc[:, 'petal_width']
pltY = data.loc[:, 'species']
plt.scatter(pltX, pltY, color = 'pink', label = 'petal_width')

plt.legend(loc = 4, prop = {'size':8})
plt.show()
```

3. split the data train 80% + test 20%

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

4. train model

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

5. test model

```python
predictions = model.predict(x_test)
print(predictions)

y_test
```

6. precision, recall, f1-score

```python
print(classification_report(y_test,predictions))
```

```python
print(accuracy_score(y_test, predictions))
```
